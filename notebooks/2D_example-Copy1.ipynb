```python
import nurbs
import numpy as np
import matplotlib.pyplot as plt
```

```python
poles_x = 3
poles_y = 3
degree_x = 2
degree_y = 2
knots_x = nurbs.create_knots_vector(-1, 1, degree_x, poles_x)
knots_y = nurbs.create_knots_vector(-1, 1, degree_y, poles_y)
weights = np.array([1] * (poles_x * poles_y))
```

```python
nu = nurbs.NurbsBase2D(knots_x, knots_y, weights, degree_x, degree_y)
nu.computeFirstDerivatives()
```

```python
xi = np.linspace(-1, 1, 10)
yi = np.linspace(-1, 1, 10)
test = np.array([xi, yi]).T
mat = nu.getInfluenceMatrix(test)
```

```python
x = np.linspace(0, 1, 3)
y = np.linspace(0, 1, 3)
poles = np.array(np.meshgrid(x, y)).T.reshape(-1, 2)
u = poles.T[0]**2 + poles.T[1]**2
```

```python
plt.plot(mat@u)
```

```python
du = nu.getDuMatrix(test)
plt.plot(du @ u)
```

```python
dv = nu.getDvMatrix(test)
plt.plot(dv @ u)
```
